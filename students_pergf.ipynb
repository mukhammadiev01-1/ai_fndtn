```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentsPerformance.csv")

print("Dataset shape (rows, columns):", df.shape)
print("Column name:", df.columns.tolist())

print("First 5 rows of the dataset:")
print(df.head())

print("Summary statistics:")
print(df.describe())

print("Data types and null info:")
df.info()
```

```python
# Average scores
print("Average scores by subject:")
print(df[["math score", "reading score", "writing score"]].mean())
```

```python
# Maximum scores
print("Maximum scores by subject:")
print(df[["math score", "reading score", "writing score"]].max())
```

```python
# Minimum scores
print("Minimum scores by subject:")
print(df[["math score", "reading score", "writing score"]].min())
```

```python
# Correlation
print("Correlation:")
print(df[["math score", "reading score", "writing score"]].corr())
```

```python
df["total score"] = df[
    ["math score", "reading score", "writing score"]
].sum(axis=1)

df["average score"] = df["total score"] / 3

df["passed"] = np.where(
    df["average score"] >= 50,
    "Yes",
    "No"
)

print("New columns preview (total, average, passed):")
print(df[["total score", "average score", "passed"]].head())
```

```python
test_prep_group = df.groupby("test preparation course")[
    ["math score", "reading score", "writing score"]
].mean()

print("Student score by test preparation course:")
print(test_prep_group)

# Gender-based average
gender_avg = df.groupby("gender")[
    ["math score", "reading score", "writing score"]
].mean()

print("Average score by gender:")
print(gender_avg)

# Who passed vs failed
pass_rate = df["passed"].value_counts(normalize=True) * 100

print("Pass rate:\n", pass_rate)
```

```python
# Convert scores to NumPy array
scores = df[
    ["math score", "reading score", "writing score"]
].values

# Mean and standard deviation
print("NumPy Mean:", np.mean(scores, axis=0))
print("NumPy Std:", np.std(scores, axis=0))

# Z-score normalization
z_score = (
    scores - np.mean(scores, axis=0)
) / np.std(scores, axis=0)

print("Z-scores (normalized numbers):")
print(z_score[:5])

# Average = 0
# Negative = below average
# Positive = above average
```

```python
# How many students scored above 90 average?
top_scorers = df[df["average score"] >= 90]
print("Top scored students above 90:", len(top_scorers))

# How many failed?
failed = df[df["passed"] == "No"]
print("Failed from exam:", len(failed))
```
